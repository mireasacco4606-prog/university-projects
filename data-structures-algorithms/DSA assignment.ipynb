### 1. Create two integer arrays A and B. The two arrays must contain at least 256 elements but must be of unequal size. Populate both arrays with randomly generated integers between 0 and 1024. Sort array A using Shell Sort and array B using Quick Sort.

```python
import random

A = [random.randint(0, 1024) for _ in range(257)]
B = [random.randint(0, 1024) for _ in range(300)]


def shell_sort(arr):
    n = len(arr) // 2
    flag = True
    i = 0
    while n>=1 and flag:
        flag = False
        for i in range(len(arr) - n):
            if arr[i] > arr[i+n]:
                temp = arr[i]
                arr[i] = arr[i+n]
                arr[i+n] = temp
                flag = True
        if n>1:
            n = n // 2    
    return arr


def quick_sort(arr, first, last):
    if first < last:
        pivot_pos = partition(arr, first, last)
        quick_sort(arr, first, pivot_pos-1)
        quick_sort(arr, pivot_pos+1, last)

def  partition(arr, first, last):
    pivot = arr[first]
    left = first
    right = last
    
    while left<right:

        while left <= last and arr[left] <= pivot:
            left += 1

        while right >= first and arr[right] > pivot:
            right -= 1

        if left<right:
            temp = arr[left]
            arr[left] = arr[right]
            arr[right] = temp

    temp = pivot
    arr[first] = arr[right]
    arr[right] = pivot

    return right

A = shell_sort(A)
print('size of A =', len(A))
print('sorted A', A)
print('size of B =', len(B))
quick_sort(B, 0 ,len(B)-1)
print('sorted B', B)
```

### 2. Take the two sorted arrays A and B from Question 1 above and merge into a new array C. You must do this in linear time. Note that the size of C must be size of A plus size of B. Use the merging technique used in Merge Sort.


```python
def merge(arr1, arr2):
    l1 = len(arr1)
    l2 = len(arr2)
    result = []

    i = j = 0

    while i<l1 and j<l2:
        if arr1[i]<arr2[j]:
            result.append(arr1[i])
            i += 1
        
        else:
            result.append(arr2[j])
            j += 1
    
    result.extend(arr1[i:])
    result.extend(arr2[j:])

    return result

C = merge(A, B)
print('size of C =', len(C))
print('C', C)
```

### 5. Write a program that uses an ADT Stack to evaluate arithmetic expressions in RPN format. The contents of the stack should be displayed on the screen during evaluation. The allowed arithmetic operators are +, -, x, and /.

```python
inp = "5 3 + 8 2 - x"
inp_list = inp.split()
stack = []
operations = ['+', '-', 'x', '/']
for i in inp_list:
    if i not in operations:
        stack.append(int(i))
    else:
        num2 = stack.pop()
        num1 = stack.pop()
        if i =='+':
            ans = num1 + num2
        elif i =='-':
            ans = num1 - num2
        elif i =='/':
            ans = num1 / num2
        elif i =='x':
            ans = num1 * num2
        stack.append(ans)
    print(stack)
```

### 6. Write a Boolean function that checks if a number is prime. Also implement the Sieve of Eratosthenes algorithm – two separate algorithms. Explain any optimizations made.


```python
def is_prime(num):
    i=2
    while i<=(num**0.5):
        if num%i == 0:
            return False
        i += 1
    return True
    

print(is_prime(50))
print(is_prime(97))


def sieve(n):
    list = []
    for i in range(2,n+1):
        list.append(i)
    
    for i in list:
        for j in list:
            if i != j and j%i==0:
                list.remove(j)
                
    return list


print(sieve(100))
```

### 7. Write a program that generates the Collatz sequence of all the integers in the range 2 to 512. The program will store the generated sequences in a CSV file.

```python
import csv
def collatz_seq(n):
    seq = str(n)
    while n>1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = (n * 3) + 1
        seq = seq + ' ' + str(n)
    return seq

with open('collatz_seq.csv', 'w') as collatz:

    fieldnames = ['integer','sequence']

    writer = csv.DictWriter(collatz, fieldnames = fieldnames)

    writer.writeheader()

    for i in range(2, 513):
        numbers = collatz_seq(i)
        writer.writerow({'integer' : i, 'sequence' : numbers})
    
    #the csv file is in the file on this repository
```

### 8. Write a program that finds an approximation to the square root of a given number n using an iterative numerical method such as the Newton-Raphson Method.

```python
def newt_raph(n):
    x = n
    for i in range(10):
        x = 0.5 * (x + (n/x))
    return x

print(newt_raph(45)) 
print(newt_raph(36))
print(newt_raph(70))
```

### 10. Write a recursive function that finds the largest number in a given list of integers.

```python
def find_max(list):

    if len(list) == 1:
        return list[0]
    
    leng = len(list)//2
    max1 = find_max(list[leng:])
    max2 = find_max(list[:leng])

    if max1 > max2:
        largest = max1
    else:
        largest = max2

    return largest

nums = [34, 35, 120, 100, 20, 200]
print(find_max(nums))
```

### 11. Write a function that computes cosine or sine by taking the first n terms of the appropriate (Maclaurin?) series expansion. Be careful with computing large factorials.

```python
import math

def factorial(n):
    ans = 1
    for x in range(1, n+1):
        ans = x * ans
    return ans

def addition(x, start):
    ans=0
    for i in range(start, 20, 2):
        sign = (-1) ** ((i - start) // 2)
        ans = ans + (sign * (x**i)/factorial(i))
    return ans

def sine(x):
    x = math.radians(x)
    ans = addition(x, 1)
    return ans

def cosine(x):
    x = math.radians(x)
    return addition(x, 0)

print(sine(90), ',  ',cosine(90))
```

### 12. Write a function that returns the sum of the first n numbers of the Fibonacci sequence (Wikipedia). The first 2 numbers in the sequence are 1,1, …

```python
def fibonacci(n):
    fib = [1, 1]
    for i in range(2, n):
        fib.append(fib[i-2] + fib[i-1])
    print(fib)
    return sum(fib)




print(fibonacci(20))
```
